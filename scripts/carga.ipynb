Carga de datos desde la api de consultas en el sitio web de Science For a Changing World

```python
# Importar libreias necesarias
import pandas as pd
import io
import requests
import json
```

los datos se descargan en formato geojson

```python
def read_url(url):
    response = requests.get(url)
    data = json.loads(response.content.decode('utf-8'))
    df = pd.json_normalize(data['features'])
    return df   
```

Explicación de los parámetros utilizados:

* format=geojson: se solicita la información en formato GeoJSON.
* minmagnitude=1: se solicita solo los sismos con magnitud igual o superior a 1.
* minlatitude=24.6&maxlatitude=50&minlongitude=-168.75&maxlongitude=-52.875: se define un cuadro que abarca todo EEUU, incluyendo Alaska.
* starttime=2023-01-01&endtime=2023-05-31: se solicita solo los sismos ocurridos en los últimos 5 meses del año 2023.

```python
import requests
       
url_eeuu = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&minmagnitude=1&starttime=2023-01-01&endtime=2023-05-31&maxlatitude=72.66&minlatitude=18.92&maxlongitude=-129.99&minlongitude=-172.47'
df = read_url(url_eeuu)
df.rename(columns=lambda x: x.replace('properties.', ''), inplace=True)
df.head()
```
